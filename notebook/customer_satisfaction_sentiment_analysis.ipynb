### Response Rate

Start by analyzing how many customers has filled the survey, indicated by whether the overall_rating is not blank.

### CSAT Score

Measure the customer's overall satisfaction score (CSAT) with the following formula:

$$
CSAT = \frac{\Sigma\ total\ satisfaction\ score}{number\ of\ responded\ customer \times \max\ rating}
$$

The max rating is inserted to convert the CSAT score into percentage.

CSAT score can be classified into categories based on the result. There is no absolute threshold for each categories but the following is the common threshold:

- \>= 90%: Excellent
- 75%-90%: Good
- 60-75%: Fair
- \<60%: Poor

```python
# CSAT Score
max_rating = 5
# Menghitung nilai maksimum total rating yang diperoleh dari seluruh pelanggan yang merespon
n_data = (responded_customer.shape[0] * max_rating)

# Menjumlahkan semua rating dan kemudian menghitung skor CSAT
csat_score = responded_customer['overall_rating'].sum() / n_data

print(f'Overal CSAT Score: { (csat_score * 100):.1f}%')
```

Berdasarkan skor CSAT sebesar 91,2%, tingkat kepuasan pelanggan secara keseluruhan dinilai Sangat Baik. Hal ini menunjukkan bahwa sebagian besar pelanggan sangat puas dengan sistem tiket.

Measure the satisfaction score for the following attributes:

- customer service
- features
- value for money

```python
# Satisfaction Score for Attributes

customer_service_csat = responded_customer['customer_service'].sum() / n_data
features_csat = responded_customer['features'].sum() / n_data
value_for_money_csat = responded_customer['value_for_money'].sum() / n_data

print(f'Overal CSAT Score: { (csat_score * 100):.1f}%')
print(f'Customer Service CSAT Score: { (customer_service_csat * 100):.1f}%')
print(f'Features CSAT Score: { (features_csat * 100):.1f}%')
print(f'Value for Money CSAT Score: { (value_for_money_csat * 100):.1f}%')
```

Customer Service CSAT Score: 67,3% masuk dalam kategori Cukup. Meskipun kepuasan secara keseluruhan sangat baik,Customer Service merupakan area yang dapat ditingkatkan.
Skor CSAT Fitur: 88,3% termasuk dalam kategori Baik, mendekati Sangat Baik. Pelanggan umumnya puas dengan fitur yang disediakan oleh sistem tiket.
Value for Money CSAT Score: 87,6% termasuk dalam kategori Baik. Pelanggan menganggap sistem tiket menawarkan nilai yang baik untuk uang mereka.

### NPS Score

To calculate the NPS score, first we must convert the **would_you_recommend** column into proper NPS Category based on the rating value:

* Promoter: Rating 9-10
* Passive: Rating 7-8
* Detractor: Rating < 7

```python
# Category NPS
nps_value = ['promoter', 'passive', 'detractor']
nps_condition = [responded_customer['likelihood_to_recommend'].isin([9, 10]),
                 responded_customer['likelihood_to_recommend'].isin([7, 8]),
                 responded_customer['likelihood_to_recommend'] < 7]

responded_customer['nps_category'] = np.select(nps_condition, nps_value, default='Unknown')

responded_customer.value_counts('nps_category', normalize = True)
```


Calculate the NPS Score with the following formula

$$
NPS = \frac{Promoter - Detractor}{Total\ Survey\ Responded}
$$

```python
# NPS Score
nps_agg_1 = responded_customer.value_counts('nps_category').reset_index()

nps_agg_promoter = nps_agg_1[nps_agg_1['nps_category'] == 'promoter'] ['count'].item()
nps_agg_detractor = nps_agg_1[nps_agg_1['nps_category'] == 'detractor'] ['count'].item()
nps_score = (nps_agg_promoter - nps_agg_detractor) / nps_agg_1['count'].sum()

print(f'NPS Score: { (nps_score * 100):.1f}')
```

Berdasarkan skor NPS yang dihitung sebesar 11,9, skor Net Promoter Score (NPS) termasuk dalam kategori Rata-rata dan Detractor 20% Pelanggan ini tidak puas dan kemungkinan besar akan menyebarkan hal negatif.

NPS score can be ranging from -100 (when all customers are detractor) to 100 (when all customers are promoter).

NPS Score can be classified into categories based on the following threshold:

- \>= 70: Excellent
- 50-69: Very Good
- 30-49: Good
- 0-29: Average
- \< 0: Poor

## Sentiment Analysis

Create a new dataframe with no blank overall_text.

```python
# Create new dataframe
df_review_clean = df_review.dropna(axis =0, subset = 'overall_text')[['id_survey', 'overall_text']].copy()

df_review_clean.head()
```

### Text Cleansing

In order to get more accurate sentiment, several text cleansing need to be done. However, in most of recent sentiment analysis models and algorithm, the only text cleansing needed are as follows:

* Clean double whitespace
* Clean URL/website
* Clean username (mostly in social media or digital text)

```python
import re

def cleansing_text(x):
  # clean double whitespace
  out_text = ' '.join(x.split())

  # clean url
  out_text = re.sub(r"http\S+|www\S+|https\S+", 'http', out_text)

  # clean username
  out_text = re.sub(r"@\S+", '@user', out_text)

  return(out_text)

cleansing_text(" Doesn't  dissapoint. The car       was great. It was the best car rental experiences I've had! Salute to @jone who recommend https:/rental.com")
```

```python
# apply cleansing to review
df_review_clean['clean_text'] = df_review_clean['overall_text'].apply(cleansing_text)

df_review_clean.head()
```

### Sentiment Analysis

Create a sentiment categories using algorithm of your own choice.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER
try:
    analyzer = SentimentIntensityAnalyzer()
except LookupError:
    nltk.download('vader_lexicon')
    analyzer = SentimentIntensityAnalyzer()


# Create sentiment category
def categorize_sentiment(text):
    score = analyzer.polarity_scores(text)['compound']
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# menerapkan fungsi kategorisasi pada teks yang telah dibersihkan.
df_review_clean['sentiment'] = df_review_clean['clean_text'].apply(categorize_sentiment)

df_review_clean.head()
```
